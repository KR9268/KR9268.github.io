## 개요 

* 한빛앤 코파일럿 세미나 다녀와서 정리 <br>


## 내용 정리

* `Auto complete` : 코드를 먼저 쓰면 해주는 제안을 활용가능
* `비사용자 대비 코딩시간 최대 55% 단축`
  * 두 그룹에 자바스크립트 작성을 시킨 후 비교 <br>
    [https://github.blog/2022-09-07-research-quantifying-github-copilots-impact-on-developer-productivity-and-happiness/](https://github.blog/2022-09-07-research-quantifying-github-copilots-impact-on-developer-productivity-and-happiness/)
* 시간을 단축하고 개발자의 정신적에너지를 절약하여 고부가가치에 집중하게 하기

* 보안이슈
  * 코파일럿은 `제품 개선을 위해 데이터를 수집`함(암호화 및 일부만 열람하게 한다고는 함)
  * 내 코드가 학습에 사용되는게 싫으면 `세팅에서 수집하지 않도록 할 수 있음`

* 코파일럿 X
  * ChatGPT처럼 질문할 수 있고 답변받을 수 있음
  * 대화한 내용에 대해 물어볼만한 질문도 선택지로 제시
  * 각 프로젝트의 언어(Python 등)에 맞춰 답변

* 대체제
  * `Codeium` : 개인사용 무료, Autocomplete 가능, Chat 가능
  * `Tabnine` : 무료버전 있음, Local Machine mode(보안에 좋을 듯)

* `사용 후기`
  * `Copilot` : 한달 무료 사용. Autocomplete기능 자체를 처음 써봐서 좋긴 했음<br>
                xml과 beautifulsoup관련 어려움이 있었는데 도움이 꽤 되었음
  * `Codeium` : 개인플랜 무료 사용. Chat기능을 별도로 쓸 일이 없었음<br>
                설치형이라 회사에 도입은 어려울 것 같았음(설치는 늘 승인받으라해서)
  * 직업 개발자가 아니어서인지 고급기능을 쓸 일이 없어서, `내 경우는 Autocomplete정도로 충족이 되어 Codeium을 사용하기로 함`
  * `(2024.06추가)` 요즘은 코드특화된 모델이 많이 나와서 채팅 쪽은 오히려 `코드스트랄` 등이 체감상 더 나은 것 같음
    * Codestral(Mistral Codemodel) : Mistral Lechat 접속 후 아래와 같이 모델을 변경하여 사용 가능 <br>
