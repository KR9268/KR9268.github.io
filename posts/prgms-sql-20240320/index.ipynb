## 작성답안

```python
SELECT board.TITLE, board.BOARD_ID,
       reply.REPLY_ID, reply.WRITER_ID, reply.CONTENTS, TO_CHAR(reply.CREATED_DATE, 'YYYY-MM-DD')
FROM USED_GOODS_BOARD board, USED_GOODS_REPLY reply
WHERE TO_CHAR(board.CREATED_DATE, 'YYYYMM') = '202210'
  AND board.BOARD_ID = reply.BOARD_ID
ORDER BY reply.CREATED_DATE ASC, board.TITLE ASC;
```

## 정리

* 각 테이블의 BOARD_ID 일치시키는 것을 실수하였음. 향후 동일케이스에 대해서는 고려하여 풀기
* 다중정렬<br>
> ORDER BY reply.CREATED_DATE ASC, board.TITLE ASC;<br>
앞의 컬럼일수록 정렬 우선순위를 가짐
